```python
import pyACA
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_names = ['poke_center.wav','01-D_AMairena.wav','flute-A4-96k.wav','piano.wav','sine-440-96k.wav', 'winds.wav']
fig,axs = plt.subplots(len(file_names),1,figsize=(10,10))
fig.suptitle('Onset detection Novelty Function plots using Flux method')
for idx, f in enumerate(file_names):
    f_s, x = pyACA.ToolReadAudio('../audio/'+f)
    d, t, iPeaks = pyACA.computeNoveltyFunction('Flux', x, f_s)
    axs[idx].plot(t, d)
    for peak in iPeaks:
        # Add red vertical line at each onset detected
        axs[idx].axvline(x=t[peak], ymin=0, ymax = 1, linewidth=2, color='r')
    axs[idx].set_title(f)

for ax in axs.flat:
    ax.set(xlabel='Time (s)', ylabel='Novelty Value')
fig.tight_layout()
```
