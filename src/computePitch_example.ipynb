```python
import pyACA
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_names = ['poke_center.wav','01-D_AMairena.wav','flute-A4-96k.wav','piano.wav','sine-440-96k.wav', 'winds.wav']
fig,axs = plt.subplots(len(file_names),1,figsize=(10,10))
fig.suptitle('Pitch maps for various audio files using SpectralHps')
for idx, f in enumerate(file_names):
    f_s, x = pyACA.ToolReadAudio('../audio/'+f)
    f0, t = pyACA.computePitch(cPitchTrackName='SpectralHps', x=x, f_s=f_s, iBlockLength=4096, iHopLength=2048)
    axs[idx].plot(t, f0)
    axs[idx].set_title(f)
for ax in axs.flat:
    ax.set(xlabel='Time (s)', ylabel='f0(Hz)')
fig.tight_layout()
```
