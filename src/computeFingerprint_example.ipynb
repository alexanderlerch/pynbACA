```python
import pyACA
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_names = ['poke_center.wav','01-D_AMairena.wav','flute-A4-96k.wav','piano.wav','sine-440-96k.wav', 'winds.wav']
fig,axs = plt.subplots(len(file_names),1,figsize=(10,10))
fig.suptitle('Audio Fingerprint plots for different audio files')
for idx, f in enumerate(file_names):
    f_s, x = pyACA.ToolReadAudio('../audio/'+f)
    F,t=pyACA.computeFingerprint(x, f_s)
    axs[idx].imshow(F, aspect='auto', cmap='hot', origin='lower')
    # axs[idx].set_xlabel('Time (index)')
    # axs[idx].set_ylabel('Subfingerprint Dimension')
    axs[idx].set_title(f)
for ax in axs.flat:
    ax.set(xlabel='Block Number', ylabel='Dimension')
fig.tight_layout()
```
